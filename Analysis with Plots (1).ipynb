```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
```

```python
data1=pd.read_csv('matches')
```

```python
data2=pd.read_csv('deliveries.csv.zip')
```

```python
data=data2.merge(data1,left_on='match_id',right_on='id')
```

```python
#a function which will take a batsman name as input 
#and plot a 2d plot/line plot depicting his sesaon by season performance(in terms of runs scored per season)

def plot1(batsman_name):
    mask=data['batsman']==batsman_name
    batsman_data=data[mask]
    new=batsman_data.groupby('season')['batsman_runs'].sum()
    Season=new.index
    Runs=new.values
    plt.plot(Season,Runs,color="red",marker= '*',markersize=8)
    plt.xlabel("Year")
    plt.ylabel("Runs Scored")
    plt.title("Runs per Season")
    return plt.show()

```

```python
plot1("V Kohli")
```

```python
#a function which will accept batsman name as input and outputs a pie chart suggesting
#percentage of runs scored with scoring shots.
def plot2(batsman):
    mask2=data2['batsman']==batsman
    data3=data2[mask2]
    info=data3.groupby('batsman_runs')['batsman_runs'].sum()
    shots=info.index
    runs=info.values
    plt.pie(runs,labels=shots,autopct="%0.2f%%")
    plt.title("Runs and shots")
    return plt.show()
```
